```python
listaA = [0, 1, 2, 3, 4, 5]
listaB = [1, 2, 3, 4, 5, 6]
listaC = [0.5, 0.2, 0.01, 3.889]


listab = []
resultados = []
def somase(df):
    count = 0
    for n in df['list - A']:
        listab.append(df['List - B'][count])
        if n in listab:
            lb = df[df['List - B'] == n]['List - B'].sum()
            r = resultados[lb-1] 
            resultados.append(df['List - C'][count] + r)
        else:
            resultados.append(df['List - C'][count])
            
        count += 1
somase(df)
print(resultados)
```

```python
for num in resultados:
    print(f'{num:.2f}', end=', ')
```
