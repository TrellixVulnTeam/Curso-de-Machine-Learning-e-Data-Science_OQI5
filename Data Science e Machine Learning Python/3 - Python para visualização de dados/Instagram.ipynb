```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
```
